```python
import pandas as pd
import plotly.graph_objects as go
```

```python
data = pd.read_csv("IMDB-Movie-Data.csv")
```

```python
data.head()
```

```python
actors = data["Actors"].values
actors
```

```python
idx = 1
values = []
all_actors = []
for i in actors:
    #print(str(idx) + " " + i)
    values = i.split(',')
    for j in values:
        all_actors.append(j.strip())
all_actors
```

```python
len(all_actors)
```

```python
actor_list = []
for actor in all_actors:
    if actor not in actor_list:
        actor_list.append(actor)
len(actor_list)
```

```python
actor_dict = { i : 0 for i in actor_list}
actor_dict
```

```python
import operator

for actor in all_actors:
    actor_dict[actor] += 1
actor_dict

sorted_actor_dict = sorted(actor_dict, key=actor_dict.get, reverse=True)
for i in sorted_actor_dict:
    print(i, actor_dict[i])
```

```python
#data["Actors"].values
actors = data["Actors"].values
actors
idx = []
count = 0
for lst in actors:
    if "Mark Wahlberg" in lst:
        print(lst)
        idx.append(count)
    count += 1
idx        
```

```python
mw_data = data.iloc[idx]
mw_data.head()
```

```python
hover_text = mw_data.Title
bubble_size = mw_data['Revenue (Millions)'] / 3
fig = go.Figure(data=[go.Scatter(
    x=mw_data['Year'] ,
    y=mw_data['Revenue (Millions)'],
    mode='markers',
    text = hover_text,
    marker=dict(
        size=bubble_size,
        )
)])

fig.update_layout(
    title='Revenue for Mark Wahlberg Films',
    xaxis=dict(
        title='Year',
    ),
    yaxis=dict(
        title='Revenue (Millions)',
    ),
    showlegend=True,
)

fig.show()
```
